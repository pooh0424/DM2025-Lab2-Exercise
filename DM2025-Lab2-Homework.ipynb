## 1. Preprocessing Steps

````python
import os
import json
import sys
import pandas as pd
import numpy as np
import argparse
from typing import Optional

# Scikit-learn imports
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from scipy.special import softmax

def check_requirements():
    try:
        import transformers
        import datasets
        import torch
        return True, ''
    except Exception as e:
        msg = (
            "Missing required packages.\n"
            "Install with: pip install transformers datasets accelerate sentencepiece torch\n"
            f"Error: {e}"
        )
        return False, msg

ok, msg = check_requirements()
if not ok:
    print(msg)
else:
    print("Environment check passed.")
    from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer, DataCollatorWithPadding
    import torch

base = os.getcwd()  
kaggle_dir = os.path.join(base, 'Kaggle')
model_output_dir = os.path.join(base, 'models', 'transformer-finetuned')

json_path = os.path.join(kaggle_dir, 'final_posts.json')
emotion_csv = os.path.join(kaggle_dir, 'emotion.csv')
id_csv = os.path.join(kaggle_dir, 'data_identification.csv')
sample_sub = os.path.join(kaggle_dir, 'samplesubmission.csv')
llm_sub_path = os.path.join(kaggle_dir, 'submission_gemini_batch.csv')


def load_posts(json_path):
    posts = []
    if not os.path.exists(json_path):
        print(f"Warning: {json_path} not found.")
        return pd.DataFrame(columns=['id', 'text'])
        
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    for item in data:
        try:
            p = item['root']['_source']['post']
            posts.append({'id': p['post_id'], 'text': p.get('text', '')})
        except Exception:
            continue
    return pd.DataFrame(posts)

print('Loading data...')
posts_df = load_posts(json_path)
emo_df = pd.read_csv(emotion_csv, dtype=str) if os.path.exists(emotion_csv) else pd.DataFrame(columns=['id', 'emotion'])
id_df = pd.read_csv(id_csv, dtype=str) if os.path.exists(id_csv) else pd.DataFrame(columns=['id', 'split'])

# Merge Logic
if not posts_df.empty and not id_df.empty:
    df = posts_df.merge(id_df, on='id', how='inner').merge(emo_df, on='id', how='left')
    
    # Split train/test
    train_df = df[(df['split'] == 'train') & (df['emotion'].notna())].copy()
    test_df = df[df['split'] == 'test'].copy()
    
    print(f'Train: {len(train_df)}, Test: {len(test_df)}')
else:
    print("Data load failed or empty. Please check your file paths.")
    train_df = pd.DataFrame()
    test_df = pd.DataFrame()
````

## 2. Feature Engineering Steps

````python
def train_tfidf_classifier(train_df):
    print('Training TF-IDF classifier...')
    X_text = train_df['text'].fillna('').astype(str)
    le = LabelEncoder()
    y = le.fit_transform(train_df['emotion'])

    vectorizer = TfidfVectorizer(max_features=20000, ngram_range=(1,2), stop_words='english')
    X_train = vectorizer.fit_transform(X_text)

    clf = LogisticRegression(max_iter=2000, solver='saga', multi_class='multinomial', class_weight='balanced')
    clf.fit(X_train, y)
    
    return vectorizer, clf, le

def build_hf_dataset(df, split_name='train'):
    try:
        from datasets import Dataset
    except Exception:
        print("Datasets library not installed.")
        return None, []
        
    sub = df[df['split'] == split_name].copy()
    sub['text'] = sub['text'].fillna('').astype(str)
    sub['label'] = sub['emotion'].fillna('')
    

    labels = sorted([l for l in sub['label'].unique() if l != ''])
    label2id = {l: i for i, l in enumerate(labels)}
    sub['label_id'] = sub['label'].map(label2id).fillna(-1).astype(int)
    
    return Dataset.from_pandas(sub.reset_index(drop=True)), labels

if not train_df.empty:
    vectorizer, clf, le = train_tfidf_classifier(train_df)
    
    X_test = vectorizer.transform(test_df['text'].fillna(''))
    probs = clf.predict_proba(X_test)
    tf_probs = probs  # shape: n_test x n_classes
    tf_labels = le.inverse_transform(np.argmax(probs, axis=1))
    tf_max_prob = np.max(probs, axis=1)
    
    tf_df = pd.DataFrame({'id': test_df['id'].values, 'tf_label': tf_labels, 'tf_conf': tf_max_prob})
    print("TF-IDF Features and Predictions generated.")
    
    hf_train_ds, hf_labels = build_hf_dataset(df, split_name='train')
    hf_test_ds, _ = build_hf_dataset(df, split_name='test')
else:
    print("Skipping Feature Engineering (No Data).")
````

## 3. LLM inference

````python
import os
import json
import time
import re
import requests
import pandas as pd
from typing import List
from dotenv import load_dotenv


env_path = os.path.join(os.getcwd(), 'config', '.env')
if os.path.exists(env_path):
    load_dotenv(dotenv_path=env_path)

GEMINI_API_KEY = os.getenv('GOOGLE_API_KEY') or os.getenv('GEMINI_API_KEY')

GEMINI_MODEL = os.getenv('GEMINI_MODEL', 'gemini-2.5-flash')
GEMINI_API_URL = os.getenv('GEMINI_API_URL')


if GEMINI_API_KEY and not GEMINI_API_URL:
    GEMINI_API_URL = f"https://generativelanguage.googleapis.com/v1beta/models/{GEMINI_MODEL}:generateContent"
    print(f"Auto-configured API URL: {GEMINI_API_URL}")

GEMINI_BATCH_SIZE = int(os.getenv('GEMINI_BATCH_SIZE', '8'))
GEMINI_REQUEST_DELAY = float(os.getenv('GEMINI_REQUEST_DELAY', '1.0'))

ALLOWED_LABELS = ['joy', 'anger', 'sadness', 'fear', 'surprise', 'disgust']


def build_few_shot_examples(train_df: pd.DataFrame, n_per_label: int = 2):
    examples = []
    for lbl in ALLOWED_LABELS:
        subset = train_df[train_df['emotion'] == lbl]
        if len(subset) == 0: continue
        samples = subset.sample(min(n_per_label, len(subset)), random_state=42)
        for _, r in samples.iterrows():
            examples.append((r['text'], lbl))
    return examples

def safe_truncate(s: str, max_chars: int = 300) -> str:
    if not isinstance(s, str): return ''
    s = s.replace('\n', ' ').strip()
    return s[:max_chars].rsplit(' ', 1)[0] + '...' if len(s) > max_chars else s

def build_batch_prompt(examples: List[tuple], items: List[dict]) -> str:
    p = 'You are an emotion classifier. Return ONLY valid JSON: an array of objects with keys "id" and "label".\n'
    p += 'Allowed labels: ' + ', '.join(ALLOWED_LABELS) + '.\n'
    p += 'Example output: [{"id":"1","label":"joy"},{"id":"2","label":"sadness"}]\n\n'
    
    if examples:
        p += 'Examples:\n'
        for t, l in examples:
            p += json.dumps({'text': safe_truncate(t, 200), 'label': l}, ensure_ascii=False) + '\n'
        p += '\n'

    p += 'Task:\n'
    for it in items:
        p += json.dumps({'id': it['id'], 'text': safe_truncate(it.get('text', ''), 300)}, ensure_ascii=False) + '\n'
    return p

def query_api_robust(api_url: str, api_key: str, prompt: str) -> str:
    headers = {'Content-Type': 'application/json'}
    params = {'key': api_key}
    
    if 'gemini' in api_url or 'generateContent' in api_url:
        # Gemini format
        body = {"contents": [{"parts": [{"text": prompt}]}]}
    else:
        # Legacy PaLM/OpenAI-like format
        body = {"prompt": {"text": prompt}}

    try:
        resp = requests.post(api_url, params=params, headers=headers, json=body, timeout=60)
        
        if resp.status_code != 200:
            return f"Error {resp.status_code}: {resp.text}"
            
        j = resp.json()
        
        # 解析 Gemini 回應
        if 'candidates' in j and j['candidates']:
            parts = j['candidates'][0].get('content', {}).get('parts', [])
            if parts:
                return parts[0].get('text', '')
            return j['candidates'][0].get('output', '') # Fallback
            
        # 解析其他格式
        if 'output' in j: return j['output']
        
        return json.dumps(j) # 回傳完整 JSON 以便除錯
        
    except Exception as e:
        return f"Exception: {str(e)}"

def extract_json_array(s: str):
    if not isinstance(s, str): return None
    try:
        cleaned = re.sub(r'```json\s*|\s*```', '', s).strip()
        m = re.search(r'\[.*\]', cleaned, re.DOTALL)
        if m:
            return json.loads(m.group(0))
        return json.loads(cleaned)
    except:
        return None

def normalize_label(lbl: str) -> str:
    l = str(lbl).strip().lower()
    for allowed in ALLOWED_LABELS:
        if allowed in l: return allowed
    return 'joy' # Default fallback


EXECUTE_LLM_BATCH = True 

if EXECUTE_LLM_BATCH and GEMINI_API_KEY:
    print(f"Starting LLM Batch Inference...")
    print(f"Target URL: {GEMINI_API_URL}")
    
    if 'train_df' not in globals() or 'test_df' not in globals():
        print("Error: train_df/test_df not found. Please run Cell 1 first.")
    else:
        examples = build_few_shot_examples(train_df, n_per_label=2)
        items = test_df[['id', 'text']].to_dict(orient='records')
        
        
        results = []
        total = len(items)
        batch_size = GEMINI_BATCH_SIZE

        print(f'Processing {total} items with batch_size={batch_size}...')

        for start in range(0, total, batch_size):
            batch = items[start:start + batch_size]
            prompt = build_batch_prompt(examples, batch)
            
            resp_text = query_api_robust(GEMINI_API_URL, GEMINI_API_KEY, prompt)
            
            arr = extract_json_array(resp_text)
            
            if arr:
                id_to_label = {str(x.get('id')): normalize_label(x.get('label', '')) for x in arr if 'id' in x}
                for it in batch:
                    lab = id_to_label.get(it['id'], 'joy')
                    results.append({'id': it['id'], 'emotion': lab})
            else:
                print(f"Batch {start} failed. Response preview: {resp_text[:100]}...")
                for it in batch: results.append({'id': it['id'], 'emotion': 'joy'})
            
            if (start // batch_size) % 5 == 0:
                print(f'Progress: {min(start+batch_size, total)}/{total}')
            
            time.sleep(GEMINI_REQUEST_DELAY)

        sub_df = pd.DataFrame(results)
        
        if os.path.exists(sample_sub):
            sample_ids = pd.read_csv(sample_sub, dtype=str)['id']
            final_sub = pd.DataFrame({'id': sample_ids})
            final_sub = final_sub.merge(sub_df, on='id', how='left')
            final_sub['emotion'] = final_sub['emotion'].fillna('joy')
        else:
            final_sub = sub_df

        final_sub.to_csv(llm_sub_path, index=False)
        print(f'Done! Saved to {llm_sub_path}')

elif not GEMINI_API_KEY:
    print("Error: GEMINI_API_KEY not set. Please set it in os.environ or .env file.")
else:
    print("Skipped. Set EXECUTE_LLM_BATCH = True to run.")
````

## 4. Model Implementation Steps

````python

class Args:
    model_name = 'distilbert-base-uncased'
    output_dir = 'models/transformer-finetuned'
    epochs = 3
    batch_size = 16
    do_train = True  
    do_eval = False
    
args = Args()

def train_transformer(args, train_ds, test_ds, labels):
    if not args.do_train:
        print("Skipping Transformer training (do_train=False).")
        return

    print("Starting Transformer training setup...")
    tokenizer = AutoTokenizer.from_pretrained(args.model_name)
    
    def tokenize_fn(examples):
        return tokenizer(examples['text'], truncation=True, max_length=256)
    
    train_tok = train_ds.map(tokenize_fn, batched=True)
    
    label2id = {l: i for i, l in enumerate(labels)}
    num_labels = len(labels)
    model = AutoModelForSequenceClassification.from_pretrained(args.model_name, num_labels=num_labels)
    
    def preprocess_labels(batch):
        lab = [label2id.get(l, -1) for l in batch.get('label', [])]
        return {'label': lab}

    train_tok = train_tok.map(preprocess_labels, batched=True)
    
    training_args = TrainingArguments(
        output_dir=args.output_dir,
        num_train_epochs=args.epochs,
        per_device_train_batch_size=args.batch_size,
        logging_steps=50,
        save_strategy='epoch'
    )
    
    data_collator = DataCollatorWithPadding(tokenizer)
    
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_tok,
        data_collator=data_collator,
    )
    
    trainer.train()
    trainer.save_model(args.output_dir)
    print(f'Model saved to {args.output_dir}')

if 'hf_train_ds' in globals() and hf_train_ds:
    train_transformer(args, hf_train_ds, hf_test_ds, hf_labels)



if os.path.exists(llm_sub_path):
    llm_df = pd.read_csv(llm_sub_path, dtype=str)
    print("Loaded LLM submission.")
else:
    print(f'LLM submission not found at {llm_sub_path}. Creating dummy LLM DF.')
    llm_df = pd.DataFrame(columns=['id', 'emotion'])

if 'tf_df' in globals():
    merged = tf_df.merge(llm_df.rename(columns={'emotion': 'llm_label'}), on='id', how='left')

    # Transformer Inference 
    transformer_labels = None
    transformer_conf = None
    transformer_probs = None
    
    model_path = args.output_dir if os.path.exists(args.output_dir) else os.path.join(base, 'models', 'transformer-small')
    
    if os.path.exists(model_path):
        print(f'Loading transformer model for inference from {model_path}...')
        try:
            tokenizer = AutoTokenizer.from_pretrained(model_path)
            model = AutoModelForSequenceClassification.from_pretrained(model_path)
            model.eval()
            
            label_list = list(le.classes_)
            texts = test_df['text'].fillna('').astype(str).tolist()
            
            batch_size = 32
            preds = []
            confs = []
            probs_list = []
            
            with torch.no_grad():
                for i in range(0, len(texts), batch_size):
                    batch_texts = texts[i:i+batch_size]
                    enc = tokenizer(batch_texts, truncation=True, max_length=256, padding=True, return_tensors='pt')
                    outputs = model(**enc)
                    logits = outputs.logits
                    probs_t = torch.softmax(logits, dim=-1).cpu().numpy()
                    idxs = probs_t.argmax(axis=1)
                    
                    preds.extend([label_list[idx] if idx < len(label_list) else label_list[0] for idx in idxs])
                    confs.extend(probs_t.max(axis=1).tolist())
                    probs_list.extend(probs_t.tolist())
            
            merged['transformer_label'] = preds
            merged['transformer_conf'] = confs
            transformer_probs = np.array(probs_list)
            print("Transformer predictions generated.")
            
        except Exception as e:
            print(f'Transformer inference failed: {e}')
    else:
        print("No Transformer model found for inference.")

    ensemble_mode = 'soft' # or 'rule'
    tf_w, tr_w, llm_w = 1.0, 1.0, 1.0

    def choose(row):
        t = row.get('transformer_label')
        tf = row.get('tf_label')
        llm = row.get('llm_label')
        tconf = float(row.get('transformer_conf', 0.0))
        tfconf = float(row.get('tf_conf', 0.0))

        if pd.notna(t) and t == tf: return t
        if pd.notna(t) and pd.notna(llm) and t == llm: return t
        if pd.notna(tf) and pd.notna(llm) and tf == llm: return tf

        if ensemble_mode == 'soft' and 'transformer_probs' in globals() and transformer_probs is not None:
            n = len(le.classes_)
            idx_row = row.name
            
            tf_p = tf_probs[idx_row]
            tr_p = transformer_probs[idx_row] if idx_row < len(transformer_probs) else np.zeros(n)
            
            llm_one = np.zeros(n)
            if pd.notna(llm) and llm in le.classes_:
                llm_one[list(le.classes_).index(llm)] = 1.0
            
            combined = tf_w * tf_p + tr_w * tr_p + llm_w * llm_one
            return le.inverse_transform([np.argmax(combined)])[0]

        if pd.notna(t) and tconf >= 0.60: return t
        if tfconf >= 0.60: return tf
        
        return llm if pd.notna(llm) else (t if pd.notna(t) else tf)

    print("Running Ensemble...")
    merged['ensemble'] = merged.apply(choose, axis=1)

    agree = (merged['tf_label'] == merged['llm_label']).sum()
    print(f"Agreement between TF-IDF and LLM: {agree}/{len(merged)}")

    if os.path.exists(sample_sub):
        sample_ids = pd.read_csv(sample_sub, dtype=str)['id'].tolist()
        final = pd.DataFrame({'id': sample_ids}).merge(merged[['id', 'ensemble']], on='id', how='left')
        most_freq = final['ensemble'].mode().iloc[0] if not final['ensemble'].dropna().empty else 'joy'
        final['emotion'] = final['ensemble'].fillna(most_freq)
        
        out_path = os.path.join(kaggle_dir, 'submission_gemini_ensemble.csv')
        final[['id', 'emotion']].to_csv(out_path, index=False)
        print(f'Saved ensemble submission to {out_path}')
    else:
        print("Sample submission file not found, skipping final CSV generation.")
````
